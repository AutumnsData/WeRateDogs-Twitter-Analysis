The wrangling efforts of the data for the WeRateDogs contest included gathering the data from multiple sources, assessing the retrieved data for cleanliness, and then cleaning the data for further analysis. 

The data for this project was gathered from a csv of the twitter archive, an image prediction tsv file hosted on Udacity’s servers, and queried from Twitter’s API. The csv was able to be read immediately into a data frame using the Pandas library. The tsv image prediction file was able to be downloaded from Udacity with the Request library. Twitter data was accessed via an API using the Tweepy library and each tweet’s ID, then gathered into a txt file with each tweet’s JSON data on its own line. This file was then able to be read into a Pandas data frame.

After all data was gathered and imported into Jupyter, the data was then assessed for required cleaning. Because this data has a large number of areas that could be cleaned, the project was limited to 8 cleaning tasks. 

Visual Assessment

Initially the data that was gathered was in multiple tables, which makes the data less efficient to work with. This means that the data needed to be merged prior to analysis. Upon visual assessment of the data gathered from Twitter, issues noticed were that there were retweets, there were non-dogs included in the data, there were multiple entries where the dog’s name was incomplete instead of NA, the columns containing dogs breeds or otherwise (‘p1’, ‘p2’, ‘p3’) do not have the same case styles throughout, and there were extraneous columns that can be deleted – ‘doggo’, ‘floofer’, ‘pupper’, ‘puppo’.

Programmatic Assessment

Upon programmatic assessment of the data gathered, it was found that the bool columns ‘p1_dog’, ‘p2_dog’, and ‘p3_dog’ converted to a float. Since these values are True/False, they should stay as bool. Upon inspection of the datatypes of each column, it was also found that the columns containing status_id, tweet_id, and user_id were int instead of strings. IDs should be strings since they aren’t used in calculations. Since the tsv contained the pictures and the csv contained the tweet IDs that could then be used to query Twitter’s API, the tables had to be merged prior to knowing whether each tweet was associated with a picture. Since an inner join was used to associate tweet IDs with pictures, it was found that some tweets did not contain pictures as there were fewer Tweets in the total dataset after the inner join between the datasets took place. 

Cleaning The Data

Prior to cleaning, a new instance of the dataset was created as a copy in order to prevent any changes to the original data.
Merging the CSV and TSV files initially took care of two issues in regard to the wrangling of this dataset. Not only did it resolve the tidiness issue of having the data for these individual dogs stored in separate tables, but it also removed any tweets that did not contain pictures. This was verified by comparing the number of tweet_ids to the number of jpg_urls in the dataset. 

Retweets were then removed by only including data in the dataframe where there was no retweeted_status_id. This is because the retweeted_status_id would only include data if the tweet in question included a reference to another tweet. All non-dogs were then removed from the dataset by only including rows where at least one of the columns ‘p1_dog’, ‘p2_dog’, and ‘p3_dog’ contained the value “True”, as this means that there was at least one dog breed found in the picture as associated with the tweet. These issues were then verified as resolved via a programmatic assessment that counted the number of each where there was a value in the retweeted_status_id column or if there was a number of columns which contained “False” in all three columns: ‘p1_dog’, ‘p2_dog’, and ‘p3_dog’. 

Since neither retweets or non-dogs were found, the next data wrangling effort included updating incomplete names and “None” values to NaN. Since there were both incomplete names and entries where the name was “None”, the entries that started with a lowercase letter were found first. This is due to the pattern that all formal names had first-letter capitalization. Any entries that started with a lowercase letter were then changed to “None” so that all entries of “None” could then be updated to NaN – which better functions as the lack of a value or a NULL within a database.

The majority of the rest of the issues resolved were related to formatting. All columns containing dog breeds or otherwise did not have the same case style. This was resolved by replacing the columns in the data frame with those columns in a lowercase format – ‘p1’, ‘p2’, ‘p3’. After merging the csv and tsv, the ‘p1_dog’, ‘p2_dog’, and ‘p3_dog’ columns were floats instead of bool. This was updated by casting the columns as type bool, then replacing the columns in the data frame with the updated columns. All columns containing IDs were updated to string types instead of floats via the same method as the bool column – ‘status_id’, ’tweet_id’, and ‘user_id’. 

The last step to wrangling this data for this project was the removal of the 'doggo', 'floofer', 'pupper', and 'puppo' columns. This was completed using the drop function on the final Pandas data frame, then verified using the columns function. After this cleaning effort, the finalized data was saved to a CSV file named "twitter_archive_master.csv".
